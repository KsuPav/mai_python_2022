Подсчитайте, сколько миссий налетал каждый пилот. Выведите результат в порядке убывания миссий

ИНФОРМАЦИЯ:

структура данных в файле: {"имя_пилота": "список_миссий":[миссия1, ...]]

структура одной миссии: {"drone":"модель_дрона", "mission":[список точек миссии]}

у пилотов неодинаковое количество миссий (и миссии могут быть разной длины). у каждой миссии - произвольная модель дрона

РЕЗУЛЬТАТ:

Пилоты в порядке убывания количества миссий: {'pilot3': 6, 'pilot8': 6, 'pilot6': 5, 'pilot2': 5, 'pilot7': 4, 'pilot9': 3, 'pilot5': 3, 'pilot4': 2, 'pilot1': 1}

```python
num_missions = {}
sorted_num_missions = {}

for pilot in pilot_path.keys():
  num_missions[pilot] = len(pilot_path[pilot]['missions'])

values = list(num_missions.values())
values.sort(reverse=True)

for val in values:
  for k, v in num_missions.items():
    if v == val:
      sorted_num_missions[k] = v
      del num_missions[k]
      break

print(f'Пилоты в порядке убывания количества миссий: {sorted_num_missions}')
```

**TODO 2-2**

Получите и выведите список всех моделей дронов, которые были в файле pilot_path.json


Подсказка: внутри print используйте str.join(), чтобы соединить элементы списка (множества)

РЕЗУЛЬТАТ:

Полеты совершались на дронах следующих моделей: DJI Mavic 2 Pro, DJI Mavic 3, DJI Inspire 2, DJI Mavic 2 Zoom, DJI Mavic 2 Enterprise Advanced

```python
set_drones = set()
for pilot in pilot_path.keys():
  for mission in pilot_path[pilot]['missions']:
    set_drones.add(mission['drone'])

print(f'Полеты совершались на дронах следующих моделей: {", ".join(set_drones)}')
```

**TODO 2-3**

Получите список миссий для каждой модели дронов, которые были в файле pilot_path.json, и выведите на экран модель дрона и количество миссий, которые он отлетал

РЕЗУЛЬТАТ:

Дрон DJI Mavic 2 Zoom отлетал 9 миссий


```python
drone_missions = {}

for drone in set_drones:
  drone_missions[drone] = 0

for pilot in pilot_path.keys():
  for i in range(sorted_num_missions[pilot]):

    drone = pilot_path[pilot]['missions'][i]['drone']
    drone_missions[drone] += 1

for k in drone_missions:
  print(f'Дрон {k} отлетал {drone_missions[k]} миссий')
```

Создайте класс MultirotorUAV - наследник классов Aircraft и UAV (лекция 7)

```python
class Aircraft:
	def __init__(self, weight):
		self._weight = weight

class UAV:
	def __init__(self):
		self._has_autopilot = True
		self._missions = []

	# напишите код для декоратора атрибута _missions
	@property
	def missions(self):
		return self._missions

	@missions.setter
	def missions(self, mission):
		self._missions.append(mission)

	# напишите публичный метод count_missions
	def count_missions(self):
		return self.missions[0]


class MultirotorUAV(Aircraft, UAV):
	def __init__(self, weight, model, brand):
		super().__init__(weight)
		UAV.__init__(self)
		self.__weight = weight
		self.__model = model
		self.__brand = brand

	# напишите публичный метод get_info
	def get_info(self):
		return f'масса {self.__weight}, производитель {self.__brand}, количество миссий {self.count_missions()}'

	# напишите публичный метод get_model
	def get_model(self):
		return self.__model
```
